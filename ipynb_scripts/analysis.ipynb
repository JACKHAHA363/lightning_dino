```python
import torch
data = torch.load('../conf_mat.pkl', map_location='cpu')
conf_mat = data['conf_mat']
classes = data['classes']
```

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
# Read labels
with open('/datasets01/imagenet_full_size/061417/labels.txt') as f:
    labels = f.readlines()
labels = [l.rstrip('\n').split(',') for l in labels]
labels_to_words = {l[0]: l[1] for l in labels}
```

```python
labels_to_words
```

```python
for cls_id in range(1000):
    topk_words = conf_mat[cls_id].topk(5).indices
    topk_words = tokenizer.convert_ids_to_tokens(topk_words)
    cls = labels_to_words[classes[cls_id]]
    print('###################')
    print("Class:", cls)
    print("Top k cluster", topk_words)
```
